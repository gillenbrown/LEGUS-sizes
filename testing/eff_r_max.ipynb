```python
%matplotlib inline
import numpy as np

import betterplotlib as bpl
bpl.set_style()
```

```python
def r_eff_over_a_no_max(eta):
    return 1 * np.sqrt(2**(1/(eta-1)) - 1)

def r_eff_over_a_with_max(eta, r_max_over_a):
    return np.sqrt((0.5 * (1 + (1 + r_max_over_a**2)**(1-eta)))**(1/(1-eta)) - 1)

# def r_eff_over_a_with_max_oleg(eta, r_max_over_a):
#     if eta == 1:
#         return 2 * np.sqrt(np.sqrt(1 + r_max_over_a**1) - 1)
#     # This is so many nested parenthesis, I'll do it piece by piece
#     return np.sqrt((0.5 * (1 + (1 + r_max_over_a**2)**(1-eta)))**(1/(1-eta)) - 1)
```

```python
["#2a9d8f", "#e9c46a", "#f4a261", "#e76f51"]
```

```python
fig, ax = bpl.subplots(figsize=[7, 7])
d_eta = 0.01
etas = np.arange(-0.5 * d_eta, 5, d_eta)

ax.plot(etas, r_eff_over_a_no_max(etas), c="#264653", label='No $R_{max}$')
for r_max, color in zip([3, 5, 10, 15], ["#e76f51", "#f4a261", "#e9c46a", "#2a9d8f"]):
    ax.plot(etas, r_eff_over_a_with_max(etas, r_max), c=color, label='$R_{max}/a$ = ' + str(r_max))

# ax.fill_between([0, 1.5], 0, 3000, color="0.9")

ax.set_limits(0, 3, 0, 11)
ax.axvline(1, ls=":")
ax.legend()
ax.add_labels("$\eta$ (EFF Power Law Slope)", "$R_{eff}/a$")
fig.savefig("r_max_affect.pdf")
```
